```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Dependencies
import json
import os

# Load JSON
filepath = os.path.join("data", "open-beer-database-breweries.json")
with open(filepath) as jsonfile:
    beer_json = json.load(jsonfile)
```

```python
beer_jsondf = pd.DataFrame(columns=['City','State','brewery_name','Country','Latitude','Longitude'])
beer_jsondf
```

```python
for data in beer_json:
    city=[]
    State=[]
    beer_id=[]
    lat=[]
    long=[]
    try:
        city = data['fields']['city']
        state = data['fields']['state']
        #df = df.append({'City':city},ignore_index=True)
        print(city)
        brewery_name = data['fields']['name_breweries']
        lat = data['fields']['coordinates'][0]
        long = data['fields']['coordinates'][1]
        country = data['fields']['country']
        print(brewery_name)
        print(lat)
        beer_jsondf = beer_jsondf.append({'City':city,'State':state,'brewery_name':brewery_name,'Latitude':lat,'Longitude':long,'Country':country},ignore_index=True)
    except (KeyError, IndexError):
         print("Missing field/result... skipping.")
    
```

```python
beer_jsondf = beer_jsondf.dropna()
```

```python
# drop duplicates
beer_json = beer_jsondf.drop_duplicates(["brewery_name"])
```
