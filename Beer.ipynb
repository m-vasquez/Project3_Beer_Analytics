```python
beer = pd.read_csv(os.path.join('data', 'beer_reviews.csv'))
beer.head()
```

```python
X = beer.drop(["brewery_name","beer_beerid","review_profilename","review_time"],axis =1)
X
```

```python
y = beer["beer_beerid"]
y
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
#label_encoder.fit(["beer_style","beer_name"])
label_encoder.fit(["beer_style"])
#encoded_X = label_encoder.transform(["beer_style","beer_name"])
encoded_X = label_encoder.transform(["beer_style"])
```

```python
for label, original_class in zip(encoded_X,X):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```
